```python
%reload_ext autoreload
%autoreload 2
import autopre as ap
import pandas as pd
```

```python
df = ap.read_data('./Data/device_prices.csv')
df.head(1)
```

```python
df = ap.drop_duplicates(df)
print(df.shape)
df.head(1)
```

```python
df = ap.check_missing_data(df)
```

```python
import re 

# extract digit from string and convert it into int 
def extract_digit(x):
    return int(re.findall(r'\d+', x)[0])

def remove_seperator(x):
    return x.replace(',', '')

df['Ram'] = df['Ram'].apply(extract_digit)
df['Battery'] = df['Battery'].apply(extract_digit)
df['Display'] = df['Display'].apply(extract_digit)
df['Price'] =  df['Price'].apply(remove_seperator).astype(int)

df.head(1)
```

```python
# drop all the rows with missing values
df.dropna(axis=0,inplace=True) 
```

```python
df.to_csv('./Data/cleaned_device_prices.csv', index=False)
```
