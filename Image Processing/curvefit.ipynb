```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline


# Sample data (modify these values with your actual coordinates)
x = np.array([1, 2, 3, 4, 5, 6])  # Can have 6 or 7 points
y = np.array([2, 3, 5, 7, 2, 1])

# Create a cubic spline object
spline = CubicSpline(x, y)

# Generate new x-values for a smoother curve (optional)
new_x = np.linspace(min(x), max(x), 100)  # Adjust the number of points for desired smoothness

# Evaluate the spline at the new x-values
new_y = spline(new_x)

# Plot the original data points and the fitted spline
plt.plot(x, y, 'o', label='Data Points')
plt.plot(new_x, new_y, label='Spline Fit')

# Customize the plot (optional)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Spline Fit')
plt.legend()
plt.grid(True)

plt.show()
```
